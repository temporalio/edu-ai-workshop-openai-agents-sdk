## üêõ Troubleshooting

<details>
<summary><strong>‚ùå Problem: Cell 1 fails with download error</strong></summary>

**Solution:** Check your internet connection and try again. The install script downloads from temporal.download.

</details>

<details>
<summary><strong>‚ùå Problem: Cell 2 shows an error immediately</strong></summary>

**Possible causes:**
1. Cell 1 didn't complete successfully - re-run Cell 1
2. Temporal is already running - check if you have another instance running

**Check if Temporal is already running:**
Open a new terminal and run:
```bash
pgrep -f "temporal server"
```

If it shows a process ID, Temporal is already running!

</details>

<details>
<summary><strong>‚ùå Problem: Can't find the Ports tab in VS Code</strong></summary>

**Solution:** 
1. Look at the bottom panel of VS Code
2. You should see tabs: "Terminal", "Problems", "Output", "Debug Console", "Ports"
3. If you don't see "Ports", you might not be in a Codespace
4. If running locally, just open http://localhost:8233 in your browser

</details>

<details>
<summary><strong>‚ùå Problem: Port 8233 doesn't appear in Ports tab</strong></summary>

**Solution:**
1. Wait 10-15 seconds after starting the server
2. Check that Cell 2 is still running (look for [*] on the left)
3. Look at the output from Cell 2 - it should show "UI: http://localhost:8233"
4. If still not visible, manually add it: Click "+" in Ports tab, enter "8233"

</details>

<details>
<summary><strong>‚ùå Problem: I accidentally stopped Cell 2</strong></summary>

**Solution:** No problem! Just run Cell 2 again. The server will restart.

</details>

<details>
<summary><strong>‚ùì Question: How do I stop the Temporal server?</strong></summary>

**When you're done with the workshop:**
1. Click on Cell 2 (the running cell)
2. Click the ‚èπÔ∏è stop button that appears on the left
3. Or: Use the "Interrupt" button in the notebook toolbar

```bash

```
